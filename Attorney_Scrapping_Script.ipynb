```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pyautogui  # Import pyautogui library

# Function to create a directory if it doesn't exist
def create_directory_if_not_exists(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Define the city name and document number
city_name = "Tiverton"
document_number = 1

# Create a directory for the city
city_directory = os.path.join(".", city_name)
create_directory_if_not_exists(city_directory)

# Initialize WebDriver
driver = webdriver.Chrome()

# Navigate to the website
driver.get("https://i2b.uslandrecords.com/RI/")
time.sleep(5)

# Select the city
select_city = driver.find_element(By.XPATH ,"//select[@id='RegistryMap1_RegistryListBox']")
select_city.send_keys(city_name)
time.sleep(5)

# Click Go button
driver.find_element(By.XPATH, "//input[@id='GoButton']").click()
time.sleep(10)

# Select search criteria
driver.find_element(By.XPATH, "//select[@id='SearchCriteriaName1_DDL_SearchName']").send_keys("Date Search")
time.sleep(5)

# Click Search button
driver.find_element(By.ID, 'SearchFormEx1_btnSearch').click()
time.sleep(10)

# Enter date
driver.find_element(By.ID, 'SearchFormEx1_DRACSTextBox_DateFrom').clear()
time.sleep(5)
driver.find_element(By.ID, 'SearchFormEx1_DRACSTextBox_DateFrom').send_keys("8/1/2023")
time.sleep(5)

# Click Advanced button
driver.find_element(By.XPATH, "//input[@id='SearchFormEx1_BtnAdvanced']").click()
time.sleep(10)

# Select page size
driver.find_element(By.XPATH, "//option[@value='-2']").click()
time.sleep(5)

# Select page size
driver.find_element(By.XPATH, "//option[@value='25']").click()
time.sleep(5)

# Click Search button
driver.find_element(By.ID, 'SearchFormEx1_btnSearch').click()
time.sleep(10)
i = 2
initial_window = driver.window_handles[0]
while True:
    if i < 10:
        extension = "0" + str(i)
    else:
        extension = str(i)
    driver.switch_to.window(initial_window)
    image_icon_xpath = f"//input[@id='DocList1_GridView_Document_ctl{extension}_ImgBut']"
    image_icon = driver.find_elements(By.XPATH, image_icon_xpath)
    if not image_icon:
        break
    image_icon[0].click()
    i += 1
    time.sleep(10)
    new_window = driver.window_handles[1]
    driver.switch_to.window(new_window)
    page_number = 1
    document_directory = os.path.join(city_directory, f"document_{document_number}")
    create_directory_if_not_exists(document_directory)
    while True:
        time.sleep(5)
        driver.find_element(By.ID, 'ImageViewer1_BtnFitToHeight').click()
        driver.find_element(By.ID, 'ImageViewer1_BtnFitToBest').click()
#         driver.execute_script("window.scrollBy(0,500)");
        time.sleep(5)
    
        # Get the current mouse position
        x, y = pyautogui.position()

        # Perform a left click and hold
        pyautogui.mouseDown(x, y, button='left')

        # Move the mouse up and down
        for _ in range(10):
            pyautogui.move(0, 10) # Move 10 pixels down
            time.sleep(0.1) # Wait for a short duration

#         pyautogui.move(0, -10) # Move 10 pixels up
#         time.sleep(0.1) # Wait for a short duration

        # Release the left click
        pyautogui.mouseUp(x, y, button='left')
        
        # After zooming in, get the element containing the image
#         image_container = driver.find_element(By.ID, 'ImageViewer1')

        # Scroll the image container downwards
#         driver.execute_script("arguments[0].scrollTop += 100", image_container)

        driver.save_screenshot(os.path.join(document_directory, f"{city_name}_d{document_number}_p{page_number}.png"))
        time.sleep(5)
        next_button = driver.find_elements(By.XPATH, '//*[@id="ImageViewer1_BtnNext"]')
        if not next_button:
            break
        next_button[0].click()
        page_number += 1
        time.sleep(10)
    document_number += 1

# Close the browser
driver.quit()
```

```python
import pyautogui
import pygetwindow as gw
import keyboard

# Disable PyAutoGUI fail-safe
pyautogui.FAILSAFE = False

# Get a list of all open windows
windows = gw.getAllTitles()

# Find the window corresponding to your browser (e.g., Google Chrome)
browser_window_title = [window for window in windows if "Welcome to 20/20 Perfect Vision Land Records I2: ImageViewer - Google Chrome" in window]

# Focus on the browser window
if browser_window_title:
    browser_window = gw.getWindowsWithTitle(browser_window_title[0])[0]
    browser_window.activate()
else:
    print("Browser window not found.")

# Wait for the browser window to be active
pyautogui.PAUSE = 1

# Get the position of the browser window
browser_x, browser_y, _, _ = browser_window.box

## Define the safe area to drag the mouse within
safe_area_top = browser_y + 100
safe_area_bottom = browser_y + browser_window.height - 100

# Move the mouse to the bottom of the safe area
pyautogui.moveTo(browser_x + browser_window.width // 2, safe_area_bottom, duration=0.1)

# Hold down the left mouse button
pyautogui.mouseDown()

# Scroll speed
scroll_speed = 500

# Loop to scroll
while True:
    # Check if the 'Esc' key is pressed
    if keyboard.is_pressed('Esc'):
        print("Program terminated.")
        break
    
    # Check if the mouse cursor is near the top of the safe area
    if safe_area_top - 100 <= pyautogui.position().y <= safe_area_top:
        # Release the left mouse button
        pyautogui.mouseUp()
        # Move the mouse to the middle top of the screen
        pyautogui.moveTo(browser_x + browser_window.width // 2, safe_area_top + 50, duration=0.1)
        # Move the mouse back to the bottom of the safe area
        pyautogui.moveTo(browser_x + browser_window.width // 2, safe_area_bottom, duration=0.1)
        # Hold down the left mouse button
        pyautogui.mouseDown()
    
    # Scroll up
    pyautogui.moveTo(pyautogui.position().x, max(pyautogui.position().y - scroll_speed, safe_area_top), duration=0.1)
    
    
    # Print debug information
    print("Mouse Y Position:", pyautogui.position().y)
    print("Browser Y Coordinate:", browser_y)
    
    # Check if the mouse cursor has reached the bottom of the browser window
    if pyautogui.position().y <= browser_y:
        print("End of page.")
        break
```
